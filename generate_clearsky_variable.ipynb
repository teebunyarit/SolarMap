```python
import pandas as pd
import os
import numpy as np
from CUEESolar.clearskycal import calsolar
```

```python
META_FOLDER = 'C:\\Users\\Tee\\senior_project\\src\\Data_preparetion\\metadata'
```

# input of clear sky model

```python
plant =pd.read_csv(os.path.join(META_FOLDER,'plant_metadata.csv')) # position
plant
```

```python
df = pd.date_range(start='2022-02-05', end='2023-02-06',freq='30T') # time
df = pd.DataFrame(index=df)
df = df.between_time('6:00','18:00')
df.index.name = 'Datetime'
df
```

## airmass calculation

```python
a = calsolar.cal_airmass(df,plant.lat.values,plant.lng.values)
data=[]
for i in range(56):
    for j in range(len(a)):
        data.append({'Datetime':df.index[j],'site_name':plant.site_name[i],'am':a[j][i]})
a = pd.DataFrame(data)
a = a.set_index('Datetime')
a.index = a.index.tz_localize('Asia/Bangkok')
```

```python
a
```

## Zenith angle calculation

```python
b = calsolar.cal_zna(df,plant.lat.values,plant.lng.values)
data=[]
for i in range(56):
    for j in range(len(b)):
        data.append({'Datetime':df.index[j],'site_name':plant.site_name[i],'zna':b[j][i]})
b = pd.DataFrame(data)
b = b.set_index('Datetime')
b.index = b.index.tz_localize('Asia/Bangkok')
```

```python
b
```

## Clear sky irradiance calculation

```python
c = calsolar.cal_clearsky(df,plant.lat.values,plant.lng.values,plant.lng.values)
data=[]
for i in range(56):
    for j in range(len(c)):
        data.append({'Datetime':df.index[j],'site_name':plant.site_name[i],'Iclr':c[j][i]})
c = pd.DataFrame(data)
c = c.set_index('Datetime')
c.index = c.index.tz_localize('Asia/Bangkok')
```

```python
c
```

## Merging

```python
df = c.copy()
df['am'] = a.am
df['zna'] = b.zna
df
```
