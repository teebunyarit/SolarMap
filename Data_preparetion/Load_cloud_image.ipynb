```python
import wget
import pandas as pd
import requests
import os
import urllib3
import ssl
from datetime import datetime, timedelta
ssl._create_default_https_context = ssl._create_unverified_context
urllib3.disable_warnings()
```

06:00 (UTC+7) >> 23.00 (UTC+0)  
19:00 (UTC+7) >> 12.00 (UTC+0)  
23 00 01 02 03 04 05 06 07 08 09 10 11 12 (are selected)  
13 - 23 are  night time in Thailand

```python
start_date='20220204 23:50:00'
end_date='20230206 00:00:00'
TIME = pd.date_range(start_date,end_date,freq="10min")
TIME=TIME[:-1]
TIME = TIME[(TIME.hour<13) | (TIME.hour == 23)]
```

```python
TIME
```

```python
datetime_list=[d.strftime('%Y%m%d%H%M') for d in TIME]
len(datetime_list)
```

```python
for single_datetime in datetime_list:
    site_url = 'https://161.200.85.31/api/downloadFile/ir_cloud_day/himawari_ahi_overview_'+single_datetime+'.tif'
    r = requests.head(site_url,verify=False)
    if r.status_code == 200 :
      file_name = wget.download(site_url)
      print(file_name)
    else:
      print('File:'+single_datetime+' is missing')
```

# Avialabel cloud image checking

```python
data_cm = []
```

```python
for single_datetime in datetime_list:
    print(round(datetime_list.index(single_datetime)/len(datetime_list)*100,2), end='\r')
    site_url = 'https://161.200.85.31/api/downloadFile/ir_cloud_day/himawari_ahi_overview_'+single_datetime+'.tif'
    r = requests.head(site_url,verify=False)
    time = datetime.strptime(single_datetime, '%Y%m%d%H%M')+ timedelta(hours=7)
    if r.status_code == 200 :
        data_cm.append({'Datetime': time, 'status':1})
    else:
        data_cm.append({'Datetime': time, 'status':0})
```

```python
dd1 = pd.DataFrame(data_cm)
dd1.to_csv('check_avialabel_cloudmask.csv')
```
